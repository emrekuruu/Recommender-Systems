# Content Based Recommender

```python
import pandas as pd
import numpy as np
import math
```

__Load the Data and seperate it into usable Data Frames__

```python
entire_df = pd.read_csv("content_based.csv")
```

```python
user_df = entire_df[["User 1","User 2"]][:20].T
user_df = user_df.fillna(0)
user_df
```

```python
movies_df = entire_df.iloc[:20,1:11]
attr_df = entire_df.iloc[:20,12]
movies_df = pd.DataFrame(movies_df.values.T/np.sqrt(attr_df.values), index = movies_df.columns, columns = movies_df.index).T
movies_df
```

```python
def getUserTaste(user,movie):
    return pd.DataFrame(np.dot(user,movie),index = user.index, columns = movie.columns)
```

```python
getUserTaste(user_df,movies_df)
```

```python
def similarity(user,movie):
    user_taste = getUserTaste(user,movie)
    return pd.DataFrame(np.dot(user_taste,movie.T),index = user.index, columns = movie.index + 1)+2.5
```

```python
predictions = similarity(user_df,movies_df).T
user_1 = predictions["User 1"].sort_values(ascending = False)
user_2 = predictions["User 2"].sort_values(ascending = False)
user_1
```
